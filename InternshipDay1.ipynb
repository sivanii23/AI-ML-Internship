```python
import pandas as pd
df=pd.read_csv("/content/titanic uncleaned.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoded_data=df.copy()
```

```python
label_encoder=LabelEncoder()
encoded_data['Sex_Label']=label_encoder.fit_transform(encoded_data['Sex'])
```

```python
one_hot_encoded=pd.get_dummies(encoded_data['Embarked'],prefix='Embarked') #creates dummy variables
encoded_data=pd.concat([encoded_data,one_hot_encoded],axis=1) #axis=1 means y axis
```

```python
pclass_mapping={1:'First',2:'Second',3:'Third'}
encoded_data['Pclass_Category']=encoded_data['Pclass'].map(pclass_mapping)
```

```python
print(encoded_data[['Sex', 'Sex_Label', 'Embarked', 'Embarked_Q', 'Embarked_S', 'Embarked_C', 'Pclass', 'Pclass_Category']].head())
```
