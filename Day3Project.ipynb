```python

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x=df.drop(columns=['Weekly_Sales','Date'])
y=df["Weekly_Sales"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
r2=r2_score(y_test,y_pred)
rmse = mean_squared_error(y_test, y_pred) ** 0.5

print(f'R2 Score: {r2:.4f}')
print(f'RMSE: {rmse:.4f}')
```

```python
from matplotlib import lines
plt.figure(figsize=(8,6))
sns.scatterplot(x=y_test,y=y_pred,alpha=0.7)
plt.xlabel("Actual Prices (Weekly_Sales)")
plt.ylabel("Predicted Prices (Weekly_Sales)")
plt.title("Actual vs Predicted Weekly Sales")
plt.axline([0,0],[1,1],color="red",linestyle="--")
plt.show()
```

```python
from scipy.stats import skew
```

```python
numerical_df=df.select_dtypes(include=['number'])
skewness=numerical_df.skew()
num_cols=len(numerical_df.columns)
plt.figure(figsize=(15,num_cols*3))
for i, col in enumerate(numerical_df.columns):
  plt.subplot(num_cols,2,2*i+1)
  sns.histplot(numerical_df[col],kde=True)
  plt.title(f'Histogram of {col}\nSknewness: {skewness[col]:.2f}')
  plt.subplot(num_cols,2,2*i+2)
  sns.boxplot(numerical_df[col])
  plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
print(skewness)
```

```python
import numpy as np
```

```python
skewed_cols=numerical_df.skew().abs()
highly_skewed=skewed_cols[skewed_cols>1].index
negatively_skewed=skewed_cols[skewed_cols<1].index
print("Highly Skewed Columns:\n",highly_skewed)
print("Negatively Skewed Columns:\n",negatively_skewed)
```

all values less than 1 is negatively skewed
all values greater than 1 is positively skewed

```python
numerical_df[highly_skewed]=numerical_df[highly_skewed].apply(lambda x: np.log1p(x))
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer(method='yeo-johnson')
numerical_df[negatively_skewed]=pt.fit_transform(numerical_df[negatively_skewed])
```

```python
skewness=numerical_df.skew()
print(skewness)
```

```python
x=numerical_df.drop(columns=['Weekly_Sales'])  #define features
y=numerical_df["Weekly_Sales"]   #define target
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
r2=r2_score(y_test,y_pred)
rmse = mean_squared_error(y_test, y_pred) ** 0.5

print(f"R² Score: {r2:.4f}") #performance rate
print(f"RMSE: {rmse:.4f}") #error rate
```

```python
from matplotlib.lines import lineStyles
plt.figure(figsize=(8,6))
sns.scatterplot(x=y_test,y=y_pred,alpha=0.7)
plt.xlabel("Actual Prices (Weekly_scales)")
plt.ylabel("Predicted Prices (Weekly Sales)")
plt.title("Actual vs Predicted Weekly Scales")
plt.axline([0,0],[1,1],color="red",linestyle="--")
plt.show()
```

```python
df['Store']=standard_scaler.fit_transform(df[['Store']])
df['Temperature']=standard_scaler.fit_transform(df[['Temperature']])
df['Fuel_Price']=standard_scaler.fit_transform(df[['Fuel_Price']])
df['CPI']=standard_scaler.fit_transform(df[['CPI']])
df['Unemployment']=standard_scaler.fit_transform(df[['Unemployment']])
df['Month']=standard_scaler.fit_transform(df[['Month']])
df['Day']=standard_scaler.fit_transform(df[['Day']])
```

```python
x=df.drop(columns=['Weekly_Sales','Date'])  #define features
y=df["Weekly_Sales"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
lr_r2=r2_score(y_test,y_pred)
#rmse = mean_squared_error(y_test, y_pred, squared=False)
lr_rmse = mean_squared_error(y_test, y_pred) ** 0.5

print(f"R² Score: {lr_r2:.4f}") #performance rate
print(f"RMSE: {lr_rmse:.4f}") #error rate
```

```python
from matplotlib.lines import lineStyles
plt.figure(figsize=(8,6))
sns.scatterplot(x=y_test,y=y_pred,alpha=0.7)
plt.xlabel("Actual Prices (Weekly_scales)")
plt.ylabel("Predicted Prices (Weekly Sales)")
plt.title("Actual vs Predicted Weekly Scales")
plt.axline([0,0],[1,1],color="red",linestyle="--")
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model=RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(x_train,y_train)
y_pred=rf_model.predict(x_test)
rf_r2=r2_score(y_test,y_pred)
rf_rmse=mean_squared_error(y_test,y_pred)**0.5
print(f'Random Forest R2 Score: {rf_r2:.4f}')
print(f'Random Forest RMSE: {rf_rmse:.4f}')
```
