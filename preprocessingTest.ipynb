```python
df['Age_mean_imputed']=df['Age'].fillna(df['Age'].mean())
df['Salary_mean_imputed']=df['Monthly_Salary'].fillna(df['Monthly_Salary'].mean())
df[['Age','Age_mean_imputed','Monthly_Salary','Salary_mean_imputed']]
```

```python
df['Age_median_imputed']=df['Age'].fillna(df['Age'].median())
df['Salary_median_imputed']=df['Monthly_Salary'].fillna(df['Monthly_Salary'].median())
df[['Age','Age_median_imputed','Monthly_Salary','Salary_median_imputed']]
```

```python
df['Department_mode_imputed']=df['Department'].fillna(df['Department'].mode()[0])
df['Education_mode_imputed']=df['Education'].fillna(df['Education'].mode()[0])
df['City_mode_imputed']=df['City'].fillna(df['City'].mode()[0])
df[['Department','Department_mode_imputed','Education','Education_mode_imputed','City','City_mode_imputed']]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoded_data=df.copy()
```

```python
label_encoder=LabelEncoder()
encoded_data['Gender_Label']=label_encoder.fit_transform(encoded_data['Gender'])
```

```python
one_hot_encoded=pd.get_dummies(encoded_data['Department'],prefix='Department')
encoded_data=pd.concat([encoded_data,one_hot_encoded],axis=1)
```

```python
P_Rating_mapping={1:'Poor',2:'Fair',3:'Good',4:'Very Good',5:'Excellent'}
encoded_data['P_Rating_Category']=encoded_data['Performance_Rating'].map(P_Rating_mapping)
```

```python
print(encoded_data[['Gender','Gender_Label','Department','Department_0','Department_1','Department_2','Department_3','Department_4','Performance_Rating','P_Rating_Category']].head())
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
numerical_columns=['Age','Monthly_Salary']
```

```python
encoded_data.columns
```

```python
preprocess_data=df.copy()
```

```python
preprocess_data.dropna(subset=numerical_columns,inplace=True)
```

```python
minmax_scaler=MinMaxScaler()
standard_scaler=StandardScaler()
```

```python
preprocess_data['Age_MinMax']=minmax_scaler.fit_transform(preprocess_data[['Age']])
preprocess_data['Monthly_Salary_MinMax']=minmax_scaler.fit_transform(preprocess_data[['Monthly_Salary']])
```

```python
preprocess_data['Age_MinMax']=standard_scaler.fit_transform(preprocess_data[['Age']])
preprocess_data['Monthly_Salary_MinMax']=standard_scaler.fit_transform(preprocess_data[['Monthly_Salary']])
```
